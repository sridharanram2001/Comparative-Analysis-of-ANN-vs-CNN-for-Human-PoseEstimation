```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/dataset_stick_figure9.csv')
```

```python
df.isnull().sum()
```

```python
df['pose'] = df['y'].map({"c":0,"r":1,"n":2,"q":3})
```

# Correlation plot between key points and pose

```python
plt.figure(figsize=(15,7))
df.corr()['pose'].sort_values()[:-1].plot(kind='bar')
```

# Scatterplot of x25 vs y25 keypoints (left_knee) for all the poses

```python
sns.scatterplot(x='x25',y='y25',data=df,hue='pose',palette='Dark2')
```

# Scatterplot of x11 vs y11 keypoints (left_shoulder) for all the poses

```python
sns.scatterplot(x='x11',y='y11',data=df,hue='pose',palette='Dark2')
```

# Boxplot of x25 key points for different poses

```python
sns.boxplot(x='pose',y='x25',data=df,hue='pose')
```

# Boxplot of y25 key points for different poses

```python
sns.boxplot(x='pose',y='y25',data=df,hue='pose')
```
